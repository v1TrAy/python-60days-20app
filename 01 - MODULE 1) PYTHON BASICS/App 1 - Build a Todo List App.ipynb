## Custom Input

````python
import builtins


# def input(prompt):
#     x = builtins.input(prompt)
#     print(prompt + x)
#     return x

RED = "\033[31m"
GREEN = "\033[32m"
RESET = "\033[0m"


def input(prompt):
    x = builtins.input(RED + prompt + RESET)
    print(prompt + " " + GREEN + x + RESET)
    return x
````
