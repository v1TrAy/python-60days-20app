### Todo List View and Program Exit

````python
todos = []
while True:
    user_prompt = "Type add or show or exit:"
    user_action = input(user_prompt)

    match user_action:
        case "add":
            todo = input("Enter Todo:")
            todos.append(todo)
        case "show":
            print(todos)
        case "exit":
            break
print("Cya!")
````

### Improving the Program Output

````python
todos = []
while True:
    user_prompt = "Type add or show or exit:"
    user_action = input(user_prompt)
    # Removes leading and trailing whitespace
    user_action = user_action.strip()

    match user_action:
        case "add":
            todo = input("Enter Todo:")
            todos.append(todo)
        case "show":
            for item in todos:
                print(item)
        case "exit":
            break
print("Cya!")
````

### Add an Edit Feature

````python
todos = []
while True:
    user_prompt = "Type add, edit, show or exit:"
    user_action = input(user_prompt)
    # Removes leading and trailing whitespace
    user_action = user_action.strip()

    match user_action:
        case "add":
            todo = input("Enter Todo:")
            todos.append(todo)

        case "edit":
            number = int(input("Number of the todo to edit:"))
            number = - number
            new_todo = input("Enter New Todo:")
            todos[number] = new_todo

        case "show":
            for item in todos:
                print(item)

        case "exit":
            break
print("Cya!")
````

### List VS Tuple

| Feature           | List                         | Tuple                        |
|-------------------|-------------------------------|------------------------------|
| Mutability        | Mutable (can be changed)      | Immutable (cannot be changed)|
| Syntax            | `[1, 2, 3]`                   | `(1, 2, 3)`                  |
| Methods           | Many built-in methods (e.g., `.append()`, `.remove()`) | Fewer built-in methods       |
| Performance       | Slower than tuple             | Faster than list             |
| Use Case          | When data may change          | When data should not change  |
| Memory Usage      | More memory                   | Less memory                  |
| Hashable (usable as dict key) | No                  | Yes (if all elements are hashable) |

### üß™ Exercise: Password Validation Function

**üìã Instructions**  
Complete the `strength` function. The function will check the strength of a given password. Specifically, it should:

1. Take a `password` argument,  
2. Return `"Strong Password"` if **all** of the following conditions are met:
   - The password is **8 or more characters** long,  
   - The password has at least **one uppercase letter**,  
   - The password has at least **one digit**.  
3. If **any** of the above conditions are **not met**, return `"Weak Password"`.

> üí° You may refer to the code from the Bonus Example on Day 9 as a helpful guide.

````python
def strength(password):

    result = {}

    if len(password) >= 8:
        result["length"] = True
    else:
        result["length"] = False

    digit = False
    uppercase = False

    for i in password:
        if i.isdigit():
            digit = True
        if i.isupper():
            uppercase = True

    result["digits"] = digit
    result["upper-case"] = uppercase

    if all(result.values()):
        return "Strong Password"
    else:
        return "Weak Password"
````

````python
def get_todos(filepath='todos.txt'):
    with open(filepath, 'r') as file_local:
        todos_local = file_local.readlines()
    return todos_local


def write_todos(todos_arg, filepath='todos.txt'):
    with open(filepath, 'w') as file:
        file.writelines(todos_arg)


while True:
    user_prompt = "Type add, edit, show or exit or complete:"
    user_action = input(user_prompt)
    # Removes leading and trailing whitespace
    user_action = user_action.strip()

    if user_action.startswith('add'):
        todo = user_action[4:]

        todos = get_todos()

        todos.append(todo + '\n')

        write_todos(todos)

    elif user_action.startswith('edit'):
        try:
            number = int(user_action[5:])
            number = number - 1

            todos = get_todos()

            new_todo = input("Enter New Todo:")
            todos[number] = new_todo + '\n'

            write_todos(todos)

        except ValueError:
            print("Your Command Is Not Valid!!")
            continue

    elif user_action.startswith('show'):

        todos = get_todos()

        for index, item in enumerate(todos):
            item = item.strip('\n')
            print(f"{index+1}-{item}")

    elif user_action.startswith('complete'):
        try:
            number = int(user_action[9:])

            todos = get_todos()

            index = number - 1
            todo_to_remove = todos[index].strip('\n')
            todos.pop(index)

            write_todos(todos)

            message = f'Todo {todo_to_remove} was removed from the list.'
            print(message)
        except IndexError:
            print("There is no item with that number!")
            continue

    elif user_action.startswith('exit'):
        break

    else:
        print("Command is not valid!")

print("Cya!")
````
